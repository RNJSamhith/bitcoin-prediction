```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('C:\\Users\\User\\desktop\\Datasets\\Bitcoin data.csv')
```

```python
df
```

```python
df.drop(['Date'],1,inplace=True)
df.head(7)
```

```python
prediction_days = 30
df['Prediction'] = df['Price'].shift(-prediction_days)
```

```python
df.head(7)
```

```python
df.tail(7)
```

```python
x=np.array(df.drop(['Prediction'],1))
x=x[:len(df)-prediction_days ]

print(x)
```

```python
y =np.array(df['Prediction'])
y=y[:-prediction_days ]

print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size =0.2)
```

```python
prediction_days_arr =np.array(df.drop(['Prediction'],1))[-prediction_days:]
print(prediction_days_arr)
```

```python

from sklearn.svm import SVR

svr_rbf = SVR(kernel='rbf', C=1e3, gamma=0.00001)
svr_rbf.fit(x_train, y_train)

```

```python
score=svr_rbf.score(x_test,y_test)
```

```python
svm_prediction =svr_rbf.predict(x_test)
print(svm_prediction)

print()

print(y_test)
```

```python
svm_prediction =svr_rbf.predict(prediction_days_arr)
print(svm_prediction)

print()

print(df.tail(prediction_days))
```

```python
accuracy=svr_rbf.score(x_test,y_test) 
print("Score",accuracy)
```
